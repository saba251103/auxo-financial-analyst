```python
def extract_pdf_text(file_path):
    """
    Extract text page-by-page with metadata.
    """
    doc = fitz.open(file_path)
    chunks = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        text = page.get_text()

        if text.strip():
            chunks.append({
                "source": os.path.basename(file_path),
                "page": page_num + 1,
                "content": text
            })

    return chunks
```

```python
import pandas as pd

excel_file = "/content/investor-sheet.xls"


def load_financial_sheet(sheet_name, header_row=4):
    """
    Load financial sheet while preserving hierarchy rows.
    """

    df = pd.read_excel(
        excel_file,
        sheet_name=sheet_name,
        header=header_row
    )

    # Remove junk unnamed columns
    df = df.drop(columns=['Unnamed: 0', 'Unnamed: 1'], errors='ignore')

    # Clean column names
    df.columns = df.columns.astype(str).str.strip()

    # Drop fully empty rows only
    df = df.dropna(how='all')

    # Remove footer note rows ONLY
    first_col = df.columns[0]

    df = df[
        ~df[first_col]
        .astype(str)
        .str.contains(
            "Notes|FINANCIAL STATEMENTS|Back to Index",
            case=False,
            na=False
        )
    ]

    # Preserve section hierarchy rows like:
    # Assets
    # Current assets
    # Liabilities and equity

    df = df.reset_index(drop=True)

    return df
```

```python
import pandas as pd
import re


def extract_disaggregate_revenue(file_path):
    raw = pd.read_excel(
        file_path,
        sheet_name="Disaggregate Revenue",
        header=None
    )

    revenue_tables = {}

    current_period = None
    current_rows = []

    for _, row in raw.iterrows():
        row_values = row.tolist()
        joined = " ".join([str(x) for x in row_values])

        if "Disaggregate revenue information" in joined:

            # Save previous block
            if current_period and current_rows:
                temp_df = pd.DataFrame(current_rows)

                # Use first row as header
                temp_df.columns = temp_df.iloc[0]
                temp_df = temp_df[1:]

                # Drop useless first column
                temp_df = temp_df.iloc[:, 1:]

                temp_df.columns = temp_df.columns.astype(str).str.strip()

                temp_df = temp_df.reset_index(drop=True)

                revenue_tables[current_period] = temp_df

            # Extract period
            match = re.search(r'(FY \d{2}|Q[1-4] \d{2})', joined)

            if match:
                current_period = match.group(1)

            current_rows = []

        else:
            if not all(pd.isna(x) for x in row_values):
                current_rows.append(row_values)

    # Save final block
    if current_period and current_rows:
        temp_df = pd.DataFrame(current_rows)

        temp_df.columns = temp_df.iloc[0]
        temp_df = temp_df[1:]
        temp_df = temp_df.iloc[:, 1:]
        temp_df.columns = temp_df.columns.astype(str).str.strip()
        temp_df = temp_df.reset_index(drop=True)

        revenue_tables[current_period] = temp_df

    return revenue_tables
```

```python
def chunk_text(text, chunk_size=1000, overlap=200):
    """
    Split long text into overlapping chunks.
    """

    chunks = []

    start = 0

    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]

        if chunk.strip():
            chunks.append(chunk)

        start += chunk_size - overlap

    return chunks
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

# Embedding model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Chroma client
client = chromadb.Client()
```

```python
# STEP 2  — Chunk PDFs + Batch Embed + Store in Chroma

# Build chunked documents
chunked_documents = []

for doc in all_pdf_chunks:
    text_chunks = chunk_text(doc["content"])

    for i, chunk in enumerate(text_chunks):
        chunked_documents.append({
            "id": f"{doc['source']}-{doc['page']}-{i}",
            "content": chunk,
            "source": doc["source"],
            "page": doc["page"]
        })

print("Total chunks created:", len(chunked_documents))


# Prepare batch data
documents = [doc["content"] for doc in chunked_documents]

ids = [doc["id"] for doc in chunked_documents]

metadatas = [
    {
        "source": doc["source"],
        "page": doc["page"]
    }
    for doc in chunked_documents
]


# Generate embeddings in batch
embeddings = embedding_model.encode(
    documents,
    show_progress_bar=True
).tolist()


# Reset collection safely
try:
    client.delete_collection("infosys_financial_docs")
except:
    pass

collection = client.create_collection(
    name="infosys_financial_docs"
)


# Store everything
collection.add(
    ids=ids,
    embeddings=embeddings,
    documents=documents,
    metadatas=metadatas
)

print("Stored chunks in Chroma:", collection.count())
```

```python
def classify_query(query):
    """
    Intent classifier with correct priority ordering.
    """

    q = query.lower()

    # MOST COMPLEX FIRST
    if any(word in q for word in [
        "align",
        "relationship",
        "correlate",
        "impact",
        "versus earnings",
        "vs earnings"
    ]):
        return "MULTI_HOP"

    elif any(word in q for word in [
        "summary",
        "summarize",
        "overview",
        "report"
    ]):
        return "REPORT"

    elif any(word in q for word in [
        "compare",
        "difference",
        "growth",
        "margin",
        "change"
    ]):
        return "COMPARISON"

    elif any(word in q for word in [
        "risk",
        "strategy",
        "guidance",
        "outlook",
        "commentary",
        "cyber",
        "regulation"
    ]):
        return "QUALITATIVE"

    elif any(word in q for word in [
        "stock",
        "share price",
        "market",
        "returns",
        "volatility"
    ]):
        return "STOCK"

    else:
        return "LOOKUP"
```

```python
def route_query(query):
    """
    Smarter intent routing.
    """

    q = query.lower()

    # -----------------------------------------
    # Multi-source analyst synthesis
    # -----------------------------------------
    if any(term in q for term in [
        "performance summary",
        "summarize performance",
        "market sentiment",
        "management commentary",
        "strategic risks and market reaction",
        "compare financial performance with stock",
        "compare earnings and explain",
        "financial health",
        "overall performance",
        "performance and stock",
        "compare fy25 and fy26 performance"
    ]):
        return "multi_source_report_engine"

    # -----------------------------------------
    # Hybrid reasoning
    # -----------------------------------------
    if any(term in q for term in [
        "align with earnings",
        "stock performance align",
        "market reaction",
        "strategic risks and stock"
    ]):
        return "hybrid_reasoning_engine"

    # -----------------------------------------
    # Stock
    # -----------------------------------------
    if any(term in q for term in [
        "stock",
        "share price",
        "market performance",
        "volatility",
        "return"
    ]):
        return "stock_analytics_engine"

    # -----------------------------------------
    # Qualitative / document reasoning
    # -----------------------------------------
    if any(term in q for term in [
        "strategic risks",
        "management say",
        "cybersecurity",
        "ai strategy",
        "acquire",
        "acquisition",
        "invest in",
        "bitcoin",
        "ethereum",
        "openai",
        "tesla"
    ]):
        return "vector_retrieval_engine"

    # -----------------------------------------
    # Exact comparisons
    # -----------------------------------------
    if "compare" in q:
        exact_metrics = [
            "revenue",
            "profit",
            "eps",
            "margin",
            "assets",
            "cash"
        ]

        if any(metric in q for metric in exact_metrics):
            return "financial_analytics_engine"

        return "multi_source_report_engine"

    # -----------------------------------------
    # Structured financial lookup
    # -----------------------------------------
    if any(term in q for term in [
        "revenue",
        "profit",
        "eps",
        "margin",
        "assets",
        "cash",
        "employees",
        "attrition",
        "clients"
    ]):
        return "structured_financial_engine"

    # -----------------------------------------
    # Default fallback
    # -----------------------------------------
    return "vector_retrieval_engine"
```

```python
def find_metric_row(df, metric_name):
    """
    Exact-first financial metric matching.
    Prevents wrong matches like Revenue vs Unearned Revenue.
    """

    first_col = df.columns[0]

    # Exact match first
    exact = df[
        df[first_col]
        .astype(str)
        .str.strip()
        .str.lower()
        == metric_name.lower()
    ]

    if not exact.empty:
        return exact.iloc[0]

    # Partial fallback
    partial = df[
        df[first_col]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.contains(metric_name.lower(), na=False)
    ]

    if not partial.empty:
        return partial.iloc[0]

    return None
```

```python
import pandas as pd
import numpy as np

def get_metric_value(df, row_name, period):
    first_col = df.columns[0]

    df[first_col] = df[first_col].astype(str).str.strip()

    matches = df[
        df[first_col].str.lower().str.contains(
            row_name.lower(),
            na=False
        )
    ]

    if matches.empty:
        return None

    row = matches.iloc[0]

    # exact requested period
    if period in row.index:
        val = row[period]
        if pd.notna(val):
            return val

    # fallback: latest valid numeric value
    numeric_vals = row.drop(labels=[first_col], errors="ignore")

    numeric_vals = numeric_vals[
        numeric_vals.apply(
            lambda x: pd.notna(x) and isinstance(x, (int, float, np.number))
        )
    ]

    if len(numeric_vals) > 0:
        return numeric_vals.iloc[-1]

    return None
```

```python
def compare_metrics(df, row_name, period1, period2):
    """
    Compare metric values across periods.
    """

    value1 = get_metric_value(df, row_name, period1)
    value2 = get_metric_value(df, row_name, period2)

    if value1 is None or value2 is None:
        return {
            "error": True
        }

    try:
        pct_change = ((value2 - value1) / value1) * 100
    except:
        pct_change = 0

    return {
        "period1": period1,
        "value1": value1,
        "period2": period2,
        "value2": value2,
        "pct_change": pct_change
    }
```

```python
def calculate_growth(df, metric_name, start_period, end_period):
    """
    Calculate growth percentage between two periods safely.
    """

    start_val = get_metric_value(df, metric_name, start_period)
    end_val = get_metric_value(df, metric_name, end_period)

    if start_val is None or end_val is None:
        return {
            "error": True,
            "message": "Metric lookup failed"
        }

    if start_val == 0:
        return {
            "error": True,
            "message": "Division by zero"
        }

    growth = ((end_val - start_val) / start_val) * 100

    return {
        "metric": metric_name,
        "start_period": start_period,
        "end_period": end_period,
        "start_value": start_val,
        "end_value": end_val,
        "growth_pct": growth
    }
```

```python
def get_stock_price(date, price_col="Close Price"):
    """
    Get nearest available stock price.
    """

    target_date = pd.to_datetime(date)

    eligible = stock_df[
        stock_df["Date"] <= target_date
    ]

    if eligible.empty:
        return None

    return float(eligible.iloc[-1][price_col])
```

```python
def calculate_stock_return(start_date, end_date):
    """
    Calculate stock return between two dates.
    """

    df = stock_df[
        (stock_df["Date"] >= pd.to_datetime(start_date)) &
        (stock_df["Date"] <= pd.to_datetime(end_date))
    ].copy()

    if df.empty:
        return {
            "error": True,
            "message": "No stock data found"
        }

    start_price = df.iloc[0]["Close Price"]
    end_price = df.iloc[-1]["Close Price"]

    return_pct = (
        (end_price - start_price) / start_price
    ) * 100

    return {
        "error": False,
        "start_price": float(start_price),
        "end_price": float(end_price),
        "return_pct": float(return_pct)
    }
```

```python
def calculate_stock_volatility(start_date, end_date):
    """
    Calculate daily return volatility.
    """

    df = stock_df[
        (stock_df["Date"] >= pd.to_datetime(start_date)) &
        (stock_df["Date"] <= pd.to_datetime(end_date))
    ].copy()

    if len(df) < 2:
        return None

    df["daily_return"] = df["Close Price"].pct_change()

    volatility = df["daily_return"].std() * 100

    return float(volatility)
```

```python
def calculate_max_drawdown(start_date, end_date):
    """
    Calculate max drawdown.
    """

    df = stock_df[
        (stock_df["Date"] >= pd.to_datetime(start_date)) &
        (stock_df["Date"] <= pd.to_datetime(end_date))
    ].copy()

    if df.empty:
        return None

    df["running_max"] = df["Close Price"].cummax()

    df["drawdown"] = (
        (df["Close Price"] - df["running_max"])
        / df["running_max"]
    ) * 100

    return float(df["drawdown"].min())
```

```python
def extract_stock_period(query):
    """
    Map financial periods to stock date ranges.
    """

    periods = extract_periods(query)

    mapping = {
        "FY 26": ("2025-04-01", "2026-03-31"),
        "FY 25": ("2024-04-01", "2025-03-31"),
        "FY 24": ("2023-04-01", "2024-03-31"),
        "FY 23": ("2022-04-01", "2023-03-31"),
    }

    if periods:
        return mapping.get(
            periods[0],
            ("2025-04-01", "2026-03-31")
        )

    return ("2025-04-01", "2026-03-31")
```

```python
def earnings_event_analysis(event_date, window=5):
    """
    Analyze stock movement around an event date.
    """

    event_date = pd.to_datetime(event_date)

    before = stock_df[
        stock_df["Date"] < event_date
    ].tail(window)

    after = stock_df[
        stock_df["Date"] > event_date
    ].head(window)

    if before.empty or after.empty:
        return {
            "error": True,
            "message": "Insufficient event window data"
        }

    before_price = before.iloc[-1]["Close Price"]
    after_price = after.iloc[-1]["Close Price"]

    reaction = ((after_price - before_price) / before_price) * 100

    return {
        "event_date": str(event_date.date()),
        "before_price": before_price,
        "after_price": after_price,
        "reaction_pct": reaction
    }
```

```python
import re
import pandas as pd
import numpy as np

# ============================================================
# PERIOD EXTRACTION
# ============================================================
def extract_periods(query):
    q = query.upper()

    periods = []

    fy_matches = re.findall(r"FY\s*(\d{2,4})", q)
    for year in fy_matches:
        periods.append(f"FY {year[-2:]}")

    q_matches = re.findall(r"Q([1-4])\s*FY\s*(\d{2,4})", q)
    for quarter, year in q_matches:
        periods.append(f"Q{quarter} FY {year[-2:]}")

    return periods


# ============================================================
# METRIC EXTRACTION
# ============================================================
def extract_metric(query):
    q = query.lower()

    metric_patterns = {
        "Revenue": [
            "revenue", "revenues", "sales"
        ],

        "Gross Profit": [
            "gross profit"
        ],

        "Operating Profit": [
            "operating profit", "operating income"
        ],

        "EBITDA": [
            "ebitda"
        ],

        "EBIT": [
            "ebit"
        ],

        "Net Profit": [
            "net profit",
            "profit after tax",
            "pat",
            "earnings",
            "profit"
        ],

        "EPS": [
            "eps",
            "earnings per share"
        ],

        "Operating Margin": [
            "operating margin"
        ],

        "Cash": [
            "cash",
            "cash equivalents"
        ],

        "Total Assets": [
            "total assets",
            "assets"
        ],

        "Total Liabilities": [
            "total liabilities",
            "liabilities"
        ],

        "Total Equity": [
            "equity"
        ],

        "Free Cash Flow": [
            "free cash flow",
            "fcf"
        ],

        "Operating Cash Flow": [
            "operating cash flow",
            "cash from operations"
        ],

        "Employees": [
            "employees",
            "employee",
            "headcount",
            "staff"
        ],

        "Attrition": [
            "attrition",
            "voluntary attrition"
        ],

        "Client Count": [
            "clients",
            "client count",
            "active clients"
        ],

        "Digital Revenue %": [
            "digital"
        ],

        "Utilization": [
            "utilization"
        ],

        "ROE": [
            "roe",
            "return on equity"
        ],

        "Large Deal TCV": [
            "large deal",
            "tcv"
        ]
    }

    for metric, aliases in metric_patterns.items():
        for alias in aliases:
            if alias in q:
                return metric

    return None


# ============================================================
# REGISTRY
# ============================================================
METRIC_REGISTRY = {
    "Revenue": {"df": profit_loss_df, "row": "Revenues"},
    "Gross Profit": {"df": profit_loss_df, "row": "Gross Profit"},
    "Operating Profit": {"df": profit_loss_df, "row": "Operating profit"},
    "EBITDA": {"df": profit_loss_df, "row": "Operating profit"},
    "EBIT": {"df": profit_loss_df, "row": "Operating profit"},
    "Operating Margin": {"df": profit_loss_df, "row": "Margin"},
    "Net Profit": {"df": profit_loss_df, "row": "Net profit"},
    "EPS": {"df": profit_loss_df, "row": "Basic"},

    "Cash": {"df": balance_sheet_df, "row": "cash"},
    "Total Assets": {"df": balance_sheet_df, "row": "total assets"},
    "Total Liabilities": {"df": balance_sheet_df, "row": "total liabilities"},
    "Total Equity": {"df": balance_sheet_df, "row": "equity"},

    "Free Cash Flow": {"df": cash_flow_df, "row": "free"},
    "Operating Cash Flow": {"df": cash_flow_df, "row": "operating activities"},

    "Employees": {"df": operating_metrics_df, "row": "Total Employees"},
    "Attrition": {"df": operating_metrics_df, "row": "Voluntary Attrition"},
    "Client Count": {"df": operating_metrics_df, "row": "Number of active clients"},
    "Digital Revenue %": {"df": operating_metrics_df, "row": "Digital"},
    "Utilization": {"df": operating_metrics_df, "row": "Utilization"},
    "ROE": {"df": operating_metrics_df, "row": "Return on Equity"},
    "Large Deal TCV": {"df": operating_metrics_df, "row": "Large Deal"},
}


# ============================================================
# BETTER ROW MATCHER
# ============================================================
def find_metric_row(df, row_name):
    first_col = df.columns[0]

    labels = df[first_col].astype(str).str.strip()

    # EXACT
    exact = labels.str.lower() == row_name.lower()
    if exact.any():
        return df[exact].iloc[0]

    # CONTAINS
    contains = labels.str.lower().str.contains(
        row_name.lower(),
        na=False
    )
    if contains.any():
        matches = df[contains]

        # choose longest / best descriptive match
        best = matches.iloc[
            matches[first_col].astype(str).str.len().argmax()
        ]
        return best

    return None


# ============================================================
# VALUE LOOKUP
# ============================================================
def get_metric_value(df, row_name, period):
    row = find_metric_row(df, row_name)

    if row is None:
        return None

    if period in row.index:
        val = row[period]
        if pd.notna(val):
            return float(val)

    norm_period = period.replace(" ", "").upper()

    for col in row.index:
        if str(col).replace(" ", "").upper() == norm_period:
            val = row[col]
            if pd.notna(val):
                return float(val)

    numeric_values = []

    for col in row.index:
        if col == df.columns[0]:
            continue

        val = row[col]

        if pd.notna(val):
            try:
                numeric_values.append(float(val))
            except:
                pass

    if numeric_values:
        return numeric_values[-1]

    return None


# ============================================================
# COMPARISON
# ============================================================
def compare_metrics(df, row_name, p1, p2):
    v1 = get_metric_value(df, row_name, p1)
    v2 = get_metric_value(df, row_name, p2)

    if v1 is None or v2 is None:
        return {"error": True}

    pct = ((v2 - v1) / abs(v1)) * 100

    return {
        "error": False,
        "period1": p1,
        "period2": p2,
        "value1": v1,
        "value2": v2,
        "pct_change": pct
    }


# ============================================================
# MAIN ENGINE
# ============================================================
def process_financial_query(query):
    metric = extract_metric(query)
    periods = extract_periods(query)

    if metric is None:
        return {"type": "error", "message": "Metric not identified"}

    info = METRIC_REGISTRY.get(metric)

    if info is None:
        return {"type": "error", "message": "Metric not registered"}

    df = info["df"]
    row_name = info["row"]

    if len(periods) >= 2:
        comparison = compare_metrics(
            df,
            row_name,
            periods[0],
            periods[1]
        )

        return {
            "type": "financial_comparison",
            "metric": metric,
            "comparison": comparison
        }

    if len(periods) == 1:
        value = get_metric_value(df, row_name, periods[0])

        if value is None:
            return {
                "type": "error",
                "message": f"{metric} not found"
            }

        return {
            "type": "financial",
            "metric": metric,
            "period": periods[0],
            "value": value
        }

    value = get_metric_value(df, row_name, "FY 26")

    if value is None:
        return {"type": "error", "message": "No value found"}

    return {
        "type": "financial",
        "metric": metric,
        "period": "FY 26",
        "value": value
    }
```

```python
def format_response_without_llm(query, evidence):
    """
    Deterministic fallback formatter when Gemini is unavailable
    or exact analytics are preferable.
    """

    if evidence is None:
        return "No evidence available."

    evidence_type = evidence.get("type")

    # ----------------------------------
    # Error
    # ----------------------------------
    if evidence_type == "error":
        return (
            "I could not find sufficient grounded evidence "
            "to answer this confidently."
        )

    # ----------------------------------
    # Financial lookup
    # ----------------------------------
    elif evidence_type == "financial":
        metric = evidence.get("metric", "Metric")
        period = evidence.get("period", "Unknown period")
        value = evidence.get("value", "Unavailable")

        return f"{metric} for {period} was {value}."

    # ----------------------------------
    # Financial comparison
    # ----------------------------------
    elif evidence_type == "financial_comparison":
        metric = evidence.get("metric", "Metric")
        comparison = evidence.get("comparison", {})

        if comparison.get("error"):
            return "Comparison could not be completed."

        return (
            f"{metric} changed from "
            f"{comparison.get('value1')} in {comparison.get('period1')} "
            f"to {comparison.get('value2')} in {comparison.get('period2')}, "
            f"a change of {comparison.get('pct_change'):.2f}%."
        )

    # ----------------------------------
    # Stock analytics
    # ----------------------------------
    elif evidence_type == "stock":
        result = evidence.get("result", {})

        stock_return = result.get("stock_return_pct")
        volatility = result.get("volatility_pct")
        drawdown = result.get("max_drawdown_pct")
        start_price = result.get("start_price")
        end_price = result.get("end_price")

        if stock_return is None:
            return "Stock analysis could not be completed."

        response = (
            f"Infosys stock performance from "
            f"{result.get('start_date')} to {result.get('end_date')}: "
        )

        if start_price is not None and end_price is not None:
            response += (
                f"price moved from ₹{start_price:.2f} "
                f"to ₹{end_price:.2f}. "
            )

        response += f"Total return was {stock_return:.2f}%. "

        if volatility is not None:
            response += f"Volatility was {volatility:.2f}%. "

        if drawdown is not None:
            response += f"Maximum drawdown was {drawdown:.2f}%. "

        return response

    # ----------------------------------
    # Qualitative RAG
    # ----------------------------------
    elif evidence_type == "qualitative":
        chunks = evidence.get("evidence_chunks", [])

        if not chunks:
            return "No relevant qualitative evidence was found."

        summary = "Relevant evidence found from company documents:\n"

        for chunk in chunks:
            summary += (
                f"- {chunk['source']} "
                f"(page {chunk['page']})\n"
            )

        return summary

    # ----------------------------------
    # Hybrid multi-source
    # ----------------------------------
    elif evidence_type == "hybrid":
        summary = []

        # Financial
        if "financial" in evidence:
            fin = evidence["financial"]

            if fin.get("type") == "financial":
                summary.append(
                    f"{fin['metric']} for {fin['period']} "
                    f"was {fin['value']}."
                )

            elif fin.get("type") == "financial_comparison":
                comp = fin["comparison"]

                summary.append(
                    f"{fin['metric']} changed from "
                    f"{comp['value1']} in {comp['period1']} "
                    f"to {comp['value2']} in {comp['period2']} "
                    f"({comp['pct_change']:.2f}% change)."
                )

        # Stock
        if "stock" in evidence:
            stock = evidence["stock"].get("result", {})

            if stock.get("stock_return_pct") is not None:
                stock_summary = (
                    f"Infosys stock returned "
                    f"{stock['stock_return_pct']:.2f}%"
                )

                if stock.get("volatility_pct") is not None:
                    stock_summary += (
                        f" with volatility of "
                        f"{stock['volatility_pct']:.2f}%"
                    )

                if stock.get("max_drawdown_pct") is not None:
                    stock_summary += (
                        f" and maximum drawdown of "
                        f"{stock['max_drawdown_pct']:.2f}%"
                    )

                stock_summary += "."

                summary.append(stock_summary)

        # Qualitative
        if "qualitative" in evidence:
            qual = evidence["qualitative"]

            if qual.get("type") == "qualitative":
                summary.append(
                    "Relevant qualitative evidence was retrieved "
                    "from company disclosures."
                )

        if not summary:
            return (
                "Hybrid analysis completed but "
                "no grounded evidence was available."
            )

        return " ".join(summary)

    # ----------------------------------
    # Multi-source report
    # ----------------------------------
    elif evidence_type == "multi_source_report":
        return (
            "A structured multi-source financial report "
            "was successfully generated."
        )

    # ----------------------------------
    # Unknown
    # ----------------------------------
    return "Unable to generate grounded response."
```

```python
def confidence_check(evidence):
    """
    Determine if evidence is trustworthy enough.
    """

    if evidence is None:
        return False

    evidence_type = evidence.get("type")

    if evidence_type == "error":
        return False

    if evidence_type == "qualitative":
        return len(evidence.get("evidence_chunks", [])) > 0

    if evidence_type == "financial":
        value = evidence.get("value")
        return value is not None and "not found" not in str(value).lower()

    if evidence_type == "financial_comparison":
        comparison = evidence.get("comparison", {})
        return not comparison.get("error", False)

    if evidence_type == "stock":
        result = evidence.get("result", {})
        return result.get("stock_return_pct") is not None

    if evidence_type == "multi_source_report":
        return True

    if evidence_type == "hybrid":
        has_financial = (
            "financial" in evidence and
            evidence["financial"].get("type") != "error"
        )

        has_stock = (
            "stock" in evidence and
            evidence["stock"].get("type") != "error"
        )

        has_qualitative = (
            "qualitative" in evidence and
            evidence["qualitative"].get("type") != "error"
        )

        return has_financial or has_stock or has_qualitative

    return False
```

```python
def is_followup_query(query):
    """
    Detect follow-up conversational queries.
    """

    q = query.lower()

    followup_terms = [
        "compare with",
        "what about",
        "those",
        "that",
        "them",
        "summarize that",
        "compare it",
        "did it",
        "how about"
    ]

    return any(term in q for term in followup_terms)
```

```python
def decide_output_format(query, evidence):
    q = query.lower()

    if any(k in q for k in [
        "summary",
        "summarize",
        "analysis",
        "report",
        "overview",
        "strategic risks",
        "management commentary"
    ]):
        return "pdf"

    if any(k in q for k in [
        "compare",
        "quarterly",
        "trend",
        "table",
        "excel",
        "dataset"
    ]):
        return "excel"

    if evidence.get("type") in [
        "financial_comparison",
        "stock"
    ]:
        return "excel"

    if evidence.get("type") in [
        "qualitative",
        "multi_source_report",
        "hybrid"
    ]:
        return "pdf"

    return "text"
```

```python
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Table,
    TableStyle,
    PageBreak
)
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet
from openpyxl import Workbook
from openpyxl.styles import Font
from openpyxl.utils import get_column_letter
from datetime import datetime
import os


# =========================================================
# HELPERS
# =========================================================

def autosize_columns(ws):
    for col in ws.columns:
        max_len = 0
        col_letter = get_column_letter(col[0].column)

        for cell in col:
            try:
                if cell.value:
                    max_len = max(max_len, len(str(cell.value)))
            except:
                pass

        ws.column_dimensions[col_letter].width = min(max_len + 4, 100)


def extract_citations(evidence):
    """
    Extract citations from any evidence structure.
    """

    citations = []

    evidence_type = evidence.get("type")

    if evidence_type == "qualitative":
        for chunk in evidence.get("evidence_chunks", []):
            citations.append(
                (
                    chunk.get("source", "Unknown"),
                    chunk.get("page", "N/A")
                )
            )

    elif evidence_type == "multi_source_report":
        docs = evidence.get("documents", {})
        metadata = docs.get("metadata", [])

        for meta in metadata:
            citations.append(
                (
                    meta.get("source", "Unknown"),
                    meta.get("page", "N/A")
                )
            )

    elif evidence_type == "hybrid":
        qual = evidence.get("qualitative", {})
        for chunk in qual.get("evidence_chunks", []):
            citations.append(
                (
                    chunk.get("source", "Unknown"),
                    chunk.get("page", "N/A")
                )
            )

    elif evidence_type in ["financial", "financial_comparison"]:
        result = evidence.get("result", {})
        src = result.get("source")

        if src:
            citations.append((src, "N/A"))

    return citations


# =========================================================
# PDF EXPORT
# =========================================================

def generate_pdf_report(query, evidence, response_text,
                        filename=None):
    """
    Professional PDF analyst report.
    """

    if filename is None:
        filename = f"financial_report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.pdf"

    doc = SimpleDocTemplate(filename)
    styles = getSampleStyleSheet()
    elements = []

    evidence_type = evidence.get("type")

    # Title
    elements.append(
        Paragraph("Financial Analyst Report", styles["Title"])
    )
    elements.append(Spacer(1, 20))

    # Metadata
    elements.append(
        Paragraph(
            f"<b>Generated:</b> {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}",
            styles["BodyText"]
        )
    )

    elements.append(
        Paragraph(
            f"<b>User Query:</b> {query}",
            styles["BodyText"]
        )
    )

    elements.append(Spacer(1, 20))

    # Executive Summary
    elements.append(
        Paragraph("Executive Summary", styles["Heading1"])
    )

    elements.append(
        Paragraph(response_text, styles["BodyText"])
    )

    elements.append(Spacer(1, 20))

    # ---------------------------------------------------
    # Financial
    # ---------------------------------------------------
    if evidence_type == "financial":
        result = evidence.get("result", {})

        data = [["Metric", "Value"]]

        for k, v in result.items():
            data.append([k, str(v)])

        table = Table(data)

        table.setStyle(TableStyle([
            ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
            ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
            ("GRID", (0, 0), (-1, -1), 1, colors.black)
        ]))

        elements.append(
            Paragraph("Financial Data", styles["Heading2"])
        )
        elements.append(table)

    # ---------------------------------------------------
    # Financial comparison
    # ---------------------------------------------------
    elif evidence_type == "financial_comparison":
        comparisons = evidence.get("comparisons", [])

        data = [["Metric", "Period", "Value", "Source"]]

        for comp in comparisons:
            data.append([
                comp.get("metric"),
                comp.get("period"),
                str(comp.get("value")),
                comp.get("source")
            ])

        table = Table(data)

        table.setStyle(TableStyle([
            ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
            ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
            ("GRID", (0, 0), (-1, -1), 1, colors.black)
        ]))

        elements.append(
            Paragraph("Comparison Analysis", styles["Heading2"])
        )
        elements.append(table)

    # ---------------------------------------------------
    # Stock
    # ---------------------------------------------------
    elif evidence_type == "stock":
        stock = evidence.get("result", {})

        data = [["Metric", "Value"]]

        for k, v in stock.items():
            data.append([k, str(v)])

        table = Table(data)

        table.setStyle(TableStyle([
            ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
            ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
            ("GRID", (0, 0), (-1, -1), 1, colors.black)
        ]))

        elements.append(
            Paragraph("Stock Analysis", styles["Heading2"])
        )
        elements.append(table)

    # ---------------------------------------------------
    # Qualitative
    # ---------------------------------------------------
    elif evidence_type == "qualitative":
        elements.append(
            Paragraph("Document Evidence", styles["Heading2"])
        )

        for chunk in evidence.get("evidence_chunks", []):
            source = chunk.get("source")
            page = chunk.get("page")
            snippet = chunk.get("snippet", "")[:1200]

            elements.append(
                Paragraph(
                    f"<b>{source}</b> (Page {page})",
                    styles["BodyText"]
                )
            )

            elements.append(
                Paragraph(snippet, styles["BodyText"])
            )

            elements.append(Spacer(1, 10))

    # ---------------------------------------------------
    # Multi-source report
    # ---------------------------------------------------
    elif evidence_type == "multi_source_report":
        if "financial" in evidence:
            elements.append(
                Paragraph("Financial Summary", styles["Heading2"])
            )

            data = [["Metric", "Value"]]

            for k, v in evidence["financial"].items():
                data.append([k, str(v)])

            table = Table(data)

            table.setStyle(TableStyle([
                ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
                ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
                ("GRID", (0, 0), (-1, -1), 1, colors.black)
            ]))

            elements.append(table)
            elements.append(Spacer(1, 20))

        if "stock" in evidence:
            elements.append(
                Paragraph("Stock Summary", styles["Heading2"])
            )

            data = [["Metric", "Value"]]

            for k, v in evidence["stock"].items():
                data.append([k, str(v)])

            table = Table(data)

            table.setStyle(TableStyle([
                ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
                ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
                ("GRID", (0, 0), (-1, -1), 1, colors.black)
            ]))

            elements.append(table)
            elements.append(Spacer(1, 20))

    # ---------------------------------------------------
    # Citations
    # ---------------------------------------------------
    citations = extract_citations(evidence)

    if citations:
        elements.append(PageBreak())
        elements.append(
            Paragraph("Sources Referenced", styles["Heading1"])
        )

        for src, page in citations:
            elements.append(
                Paragraph(
                    f"{src} (Page {page})",
                    styles["BodyText"]
                )
            )

    doc.build(elements)

    return filename


# =========================================================
# EXCEL EXPORT
# =========================================================

def generate_excel_report(evidence, query=None, filename=None):
    """
    Universal Excel export.
    """

    if filename is None:
        filename = f"financial_analysis_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx"

    wb = Workbook()
    ws = wb.active
    ws.title = "Analysis"

    ws.append(["Financial Analyst Chatbot Export"])
    ws["A1"].font = Font(size=16, bold=True)

    if query:
        ws.append(["Query", query])

    ws.append(["Evidence Type", evidence.get("type")])
    ws.append([])

    evidence_type = evidence.get("type")

    if evidence_type == "financial":
        result = evidence.get("result", {})

        ws.append(["Metric", "Value"])

        for k, v in result.items():
            ws.append([k, v])

    elif evidence_type == "financial_comparison":
        ws.append(["Metric", "Period", "Value", "Source"])

        for comp in evidence.get("comparisons", []):
            ws.append([
                comp.get("metric"),
                comp.get("period"),
                comp.get("value"),
                comp.get("source")
            ])

    elif evidence_type == "stock":
        result = evidence.get("result", {})

        ws.append(["Metric", "Value"])

        for k, v in result.items():
            ws.append([k, v])

    elif evidence_type == "qualitative":
        ws.append(["Source", "Page", "Snippet"])

        for chunk in evidence.get("evidence_chunks", []):
            ws.append([
                chunk.get("source"),
                chunk.get("page"),
                chunk.get("snippet")
            ])

    elif evidence_type == "multi_source_report":
        if "financial" in evidence:
            ws.append(["FINANCIAL SUMMARY"])
            ws.append(["Metric", "Value"])

            for k, v in evidence["financial"].items():
                ws.append([k, v])

            ws.append([])

        if "stock" in evidence:
            ws.append(["STOCK SUMMARY"])
            ws.append(["Metric", "Value"])

            for k, v in evidence["stock"].items():
                ws.append([k, v])

            ws.append([])

        docs = evidence.get("documents", {})
        doc_texts = docs.get("documents", [])
        metadata = docs.get("metadata", [])

        if doc_texts:
            ws.append(["DOCUMENT EVIDENCE"])
            ws.append(["Source", "Page", "Snippet"])

            for doc, meta in zip(doc_texts, metadata):
                ws.append([
                    meta.get("source"),
                    meta.get("page"),
                    doc[:2000]
                ])

    autosize_columns(ws)
    wb.save(filename)

    return filename


# =========================================================
# EXPORT ROUTER
# =========================================================

def export_response(query, evidence, response_text):
    """
    Smart export routing.
    """

    if evidence.get("type") == "error":
        return {
            "output_type": "text",
            "response": response_text
        }

    output_type = decide_output_format(query, evidence)

    if output_type == "pdf":
        file_path = generate_pdf_report(
            query,
            evidence,
            response_text
        )

        return {
            "output_type": "pdf",
            "file": file_path
        }

    elif output_type == "excel":
        file_path = generate_excel_report(
            evidence,
            query=query
        )

        return {
            "output_type": "excel",
            "file": file_path
        }

    return {
        "output_type": "text",
        "response": response_text
    }
```

```python
def process_hybrid_query(query):
    """
    Dynamic hybrid multi-source reasoning.
    """

    plan = create_analysis_plan(query)

    evidence = {
        "type": "hybrid",
        "plan": plan
    }

    # Financial
    if plan["needs_financial"]:
        evidence["financial"] = process_financial_query(query)

    # Stock
    if plan["needs_stock"]:
        evidence["stock"] = process_stock_query(query)

    # RAG
    if plan["needs_rag"]:
        evidence["qualitative"] = process_qualitative_query(query)

    return evidence
```

```python
def create_analysis_plan(query):
    """
    Dynamic planner for multi-source reasoning.
    """

    q = query.lower()

    plan = {
        "needs_financial": False,
        "needs_stock": False,
        "needs_rag": False,
        "needs_comparison": False,
        "needs_report": False
    }

    financial_terms = [
        "revenue",
        "profit",
        "earnings",
        "margin",
        "cash",
        "assets",
        "equity",
        "liabilities"
    ]

    stock_terms = [
        "stock",
        "share",
        "market",
        "price",
        "return"
    ]

    qualitative_terms = [
        "risk",
        "strategy",
        "outlook",
        "commentary",
        "disclosure"
    ]

    report_terms = [
        "summarize",
        "summary",
        "report",
        "performance"
    ]

    if any(term in q for term in financial_terms):
        plan["needs_financial"] = True

    if any(term in q for term in stock_terms):
        plan["needs_stock"] = True

    if any(term in q for term in qualitative_terms):
        plan["needs_rag"] = True

    if "compare" in q:
        plan["needs_comparison"] = True
        plan["needs_financial"] = True

    if any(term in q for term in report_terms):
        plan["needs_report"] = True

    return plan
```

```python
import google.generativeai as genai
import traceback

def build_evidence_context(evidence):
    """
    Convert actual notebook evidence structures into Gemini-readable context.
    """

    context_parts = []
    evidence_type = evidence.get("type")

    # =====================================================
    # FINANCIAL
    # =====================================================
    if evidence_type == "financial":
        result = evidence.get("result", {})

        if result:
            context_parts.append(
                f"""
FINANCIAL DATA:
Metric: {result.get('metric')}
Value: {result.get('value')}
Period: {result.get('period')}
Source: {result.get('source')}
"""
            )

    # =====================================================
    # FINANCIAL COMPARISON
    # =====================================================
    elif evidence_type == "financial_comparison":
        comparisons = evidence.get("comparisons", [])

        for comp in comparisons:
            context_parts.append(
                f"""
COMPARISON DATA:
Metric: {comp.get('metric')}
Period: {comp.get('period')}
Value: {comp.get('value')}
Source: {comp.get('source')}
"""
            )

    # =====================================================
    # STOCK
    # =====================================================
    elif evidence_type == "stock":
        result = evidence.get("result", {})

        if result:
            context_parts.append(
                f"""
STOCK ANALYSIS:
Start Price: {result.get('start_price')}
End Price: {result.get('end_price')}
Return %: {result.get('return_pct')}
"""
            )

    # =====================================================
    # QUALITATIVE
    # =====================================================
    elif evidence_type == "qualitative":
        chunks = evidence.get("evidence_chunks", [])

        for chunk in chunks:
            context_parts.append(
                f"""
SOURCE: {chunk.get('source')}
PAGE: {chunk.get('page')}

CONTENT:
{chunk.get('snippet')}
"""
            )

    # =====================================================
    # MULTI SOURCE REPORT (YOUR REAL CASE)
    # =====================================================
    elif evidence_type == "multi_source_report":

        # financial
        if "financial" in evidence:
            financial = evidence["financial"]

            context_parts.append(
                f"""
FINANCIAL SUMMARY:
{financial}
"""
            )

        # stock
        if "stock" in evidence:
            stock = evidence["stock"]

            context_parts.append(
                f"""
STOCK SUMMARY:
{stock}
"""
            )

        # documents
        docs = evidence.get("documents", {})
        doc_texts = docs.get("documents", [])
        metadata = docs.get("metadata", [])

        for doc, meta in zip(doc_texts, metadata):
            context_parts.append(
                f"""
SOURCE: {meta.get('source')}
PAGE: {meta.get('page')}

DOCUMENT CONTENT:
{doc[:2500]}
"""
            )

    # =====================================================
    # HYBRID
    # =====================================================
    elif evidence_type == "hybrid":

        if "financial" in evidence:
            context_parts.append(
                f"""
FINANCIAL:
{evidence['financial']}
"""
            )

        if "qualitative" in evidence:
            for chunk in evidence["qualitative"].get(
                "evidence_chunks", []
            ):
                context_parts.append(
                    f"""
SOURCE: {chunk.get('source')}
PAGE: {chunk.get('page')}

CONTENT:
{chunk.get('snippet')}
"""
                )

    return "\n\n".join(context_parts)
```

```python
import google.generativeai as genai
import traceback


def generate_analyst_response(query, evidence):
    """
    Generate grounded Gemini analyst response.
    Returns dict:
    {
        "success": bool,
        "response": str
    }
    """

    try:
        context = build_evidence_context(evidence)

        print("DEBUG CONTEXT LENGTH:", len(context))

        if not context.strip():
            print("DEBUG EVIDENCE:", evidence)

            return {
                "success": False,
                "response": "No grounded evidence available."
            }

        prompt = f"""
You are a senior financial analyst.

Answer ONLY using the provided evidence.

STRICT RULES:
1. Every factual claim must cite source.
2. Citation format:
   [source.pdf p.X]
3. Never invent facts.
4. If evidence exists, do NOT say insufficient evidence.
5. Be structured and professional.

QUESTION:
{query}

EVIDENCE:
{context}
"""

        model = genai.GenerativeModel("gemini-2.5-flash")

        result = model.generate_content(prompt)

        text = result.text if hasattr(result, "text") else str(result)

        return {
            "success": True,
            "response": text
        }

    except Exception as e:
        traceback.print_exc()

        return {
            "success": False,
            "response": f"Gemini failed: {str(e)}"
        }
```

```python
def resolve_followup_query(query):
    """
    Resolve ambiguous follow-up queries using history.
    """

    if len(conversation_history) == 0:
        return query

    previous_query = conversation_history[-1]["query"]

    q = query.lower()

    if "compare with" in q:
        return f"{previous_query}. {query}"

    if any(word in q for word in [
        "those",
        "that",
        "them",
        "summarize"
    ]):
        return f"Context: {previous_query}. Current question: {query}"

    return query
```

```python
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Table,
    TableStyle
)
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet


def generate_pdf_report(query, evidence, response_text,
                        filename="financial_report.pdf"):
    """
    Generate professional analyst PDF.
    """

    doc = SimpleDocTemplate(filename)
    styles = getSampleStyleSheet()

    elements = []

    # Title
    elements.append(
        Paragraph("Financial Analyst Report", styles["Title"])
    )
    elements.append(Spacer(1, 20))

    # Query
    elements.append(
        Paragraph(f"<b>User Query:</b> {query}",
                  styles["BodyText"])
    )
    elements.append(Spacer(1, 20))

    # Summary
    elements.append(
        Paragraph("<b>Executive Summary</b>",
                  styles["Heading2"])
    )
    elements.append(
        Paragraph(response_text,
                  styles["BodyText"])
    )
    elements.append(Spacer(1, 20))

    # Financial metrics
    if evidence.get("type") == "financial":
        data = [
            ["Metric", "Period", "Value"],
            [
                evidence["metric"],
                evidence["period"],
                str(evidence["value"])
            ]
        ]

        table = Table(data)

        table.setStyle(TableStyle([
            ("BACKGROUND", (0, 0), (-1, 0),
             colors.grey),
            ("TEXTCOLOR", (0, 0), (-1, 0),
             colors.whitesmoke),
            ("GRID", (0, 0), (-1, -1),
             1, colors.black)
        ]))

        elements.append(
            Paragraph("<b>Key Financial Metrics</b>",
                      styles["Heading2"])
        )

        elements.append(table)

    # Comparison
    elif evidence.get("type") == "financial_comparison":
        comp = evidence["comparison"]

        data = [
            ["Metric", "Period 1", "Value 1",
             "Period 2", "Value 2", "% Change"],
            [
                evidence["metric"],
                comp["period1"],
                comp["value1"],
                comp["period2"],
                comp["value2"],
                f"{comp['pct_change']:.2f}%"
            ]
        ]

        table = Table(data)

        table.setStyle(TableStyle([
            ("BACKGROUND", (0, 0), (-1, 0),
             colors.grey),
            ("TEXTCOLOR", (0, 0), (-1, 0),
             colors.whitesmoke),
            ("GRID", (0, 0), (-1, -1),
             1, colors.black)
        ]))

        elements.append(table)

    # Qualitative sources
    elif evidence.get("type") == "qualitative":
        elements.append(
            Paragraph("<b>Source Documents</b>",
                      styles["Heading2"])
        )

        # Iterate over evidence_chunks instead of metadata
        for chunk in evidence["evidence_chunks"]:
            src = chunk.get("source", "Unknown")
            page = chunk.get("page", "N/A")

            elements.append(
                Paragraph(
                    f"{src} (page {page})",
                    styles["BodyText"]
                )
            )

    doc.build(elements)

    return filename
```

```python
import re

def extract_periods(query):
    """
    Extract all periods from query.
    """

    query = query.upper()

    matches = re.findall(r"(Q[1-4]\s?\d{2}|FY\s?\d{2})", query)

    periods = []

    for raw in matches:
        raw = raw.strip()

        if raw.startswith("FY"):
            year = raw.replace("FY", "").strip()
            periods.append(f"FY {year}")

        elif raw.startswith("Q"):
            quarter = raw[:2]
            year = raw[2:].strip()
            periods.append(f"{quarter} {year}")

    return periods
```

```python
import time
import json

test_queries = [
    # =====================================================
    # FINANCIAL LOOKUPS
    # =====================================================
    "What was revenue in FY26?",
    "What was net profit in FY26?",
    "What was EPS in FY26?",
    "What was operating margin in FY26?",
    "What was total assets in FY26?",
    "How many employees did Infosys have?",

    # =====================================================
    # COMPARISONS
    # =====================================================
    "Compare revenue FY25 vs FY26",
    "Compare net profit FY25 vs FY26",

    # =====================================================
    # STOCK
    # =====================================================
    "What was Infosys stock performance in FY26?",
    "What was stock volatility in FY26?",

    # =====================================================
    # QUALITATIVE RAG
    # =====================================================
    "What strategic risks were mentioned by Infosys?",
    "What did management say about AI strategy?",
    "What cybersecurity risks were discussed?",

    # =====================================================
    # MULTI-SOURCE / REPORT
    # =====================================================
    "Summarize FY26 performance",
    "Compare earnings and explain management commentary",
    "Compare financial performance with stock behavior",

    # =====================================================
    # FOLLOW-UP MEMORY
    # =====================================================
    "What was FY26 revenue?",
    "Compare with FY25",

    "What strategic risks were mentioned?",
    "Summarize those",

    # =====================================================
    # HALLUCINATION TESTS
    # =====================================================
    "Did Infosys invest in Bitcoin?",
    "Did Infosys acquire Tesla?",
    "What was Infosys exposure to Ethereum?",

    # =====================================================
    # EDGE CASES
    # =====================================================
    "What was profit?",
    "How many clients?",
]

print("\nSTARTING FULL SYSTEM TEST\n")

all_results = []

for idx, q in enumerate(test_queries, start=1):
    print("=" * 100)
    print(f"TEST #{idx}")
    print("QUERY:", q)

    start = time.time()

    try:
        result = process_query(q)

        elapsed = round(time.time() - start, 2)

        route = result.get("route")
        response = result.get("response")
        evidence = result.get("evidence", {})
        export = result.get("export", {})

        print("ROUTE:", route)
        print("TIME:", elapsed, "sec")
        print("EVIDENCE TYPE:", evidence.get("type"))
        print("EXPORT TYPE:", export.get("output_type"))

        # -------------------------------------------------
        # SOURCES
        # -------------------------------------------------
        print("\nSOURCES:")

        evidence_type = evidence.get("type")

        if evidence_type == "qualitative":
            for chunk in evidence.get("evidence_chunks", []):
                print(
                    f"- {chunk.get('source')} | page {chunk.get('page')}"
                )

        elif evidence_type == "multi_source_report":
            docs = evidence.get("documents", {})
            metadata = docs.get("metadata", [])

            for meta in metadata:
                print(
                    f"- {meta.get('source')} | page {meta.get('page')}"
                )

        elif evidence_type == "hybrid":
            qual = evidence.get("qualitative", {})
            for chunk in qual.get("evidence_chunks", []):
                print(
                    f"- {chunk.get('source')} | page {chunk.get('page')}"
                )

        elif evidence_type in ["financial", "financial_comparison"]:
            result_data = evidence.get("result", {})
            if result_data.get("source"):
                print(f"- {result_data.get('source')}")

        else:
            print("- No explicit sources")

        # -------------------------------------------------
        # RESPONSE
        # -------------------------------------------------
        print("\nRESPONSE:")
        print(response[:1500])

        # -------------------------------------------------
        # EXPORT FILES
        # -------------------------------------------------
        if export.get("output_type") in ["pdf", "excel"]:
            print("\nEXPORT FILE:")
            print(export.get("file"))

        all_results.append({
            "query": q,
            "route": route,
            "response": response,
            "evidence_type": evidence.get("type"),
            "export_type": export.get("output_type"),
            "export_file": export.get("file")
        })

    except Exception as e:
        print("FAILED:", str(e))

print("\nTEST COMPLETE")
```
