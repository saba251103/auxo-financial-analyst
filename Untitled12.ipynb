```python
def extract_pdf_text(file_path):
    """
    Extract text page-by-page with metadata.
    """
    doc = fitz.open(file_path)
    chunks = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        text = page.get_text()

        if text.strip():
            chunks.append({
                "source": os.path.basename(file_path),
                "page": page_num + 1,
                "content": text
            })

    return chunks
```

```python
pdf_files = [
    "infosys-ar-25.pdf",
    "ifrs-usd-press-release_q1.pdf",
    "ifrs-usd-press-release_q2.pdf",
    "ifrs-usd-press-release_q3.pdf",
    "ifrs-usd-press-release_q4.pdf"
]

all_pdf_chunks = []

for pdf in pdf_files:
    chunks = extract_pdf_text(pdf)
    all_pdf_chunks.extend(chunks)
    print(f"{pdf}: {len(chunks)} pages extracted")
```

```python
import pandas as pd

excel_file = "investor-sheet.xls"


def load_financial_sheet(sheet_name, header_row=4):
    """
    Load financial sheet while preserving hierarchy rows.
    """

    df = pd.read_excel(
        excel_file,
        sheet_name=sheet_name,
        header=header_row
    )

    # Remove junk unnamed columns
    df = df.drop(columns=['Unnamed: 0', 'Unnamed: 1'], errors='ignore')

    # Clean column names
    df.columns = df.columns.astype(str).str.strip()

    # Drop fully empty rows only
    df = df.dropna(how='all')

    # Remove footer note rows ONLY
    first_col = df.columns[0]

    df = df[
        ~df[first_col]
        .astype(str)
        .str.contains(
            "Notes|FINANCIAL STATEMENTS|Back to Index",
            case=False,
            na=False
        )
    ]

    # Preserve section hierarchy rows like:
    # Assets
    # Current assets
    # Liabilities and equity

    df = df.reset_index(drop=True)

    return df
```

```python
import pandas as pd
import re


def extract_disaggregate_revenue(file_path):
    raw = pd.read_excel(
        file_path,
        sheet_name="Disaggregate Revenue",
        header=None
    )

    revenue_tables = {}

    current_period = None
    current_rows = []

    for _, row in raw.iterrows():
        row_values = row.tolist()
        joined = " ".join([str(x) for x in row_values])

        if "Disaggregate revenue information" in joined:

            # Save previous block
            if current_period and current_rows:
                temp_df = pd.DataFrame(current_rows)

                # Use first row as header
                temp_df.columns = temp_df.iloc[0]
                temp_df = temp_df[1:]

                # Drop useless first column
                temp_df = temp_df.iloc[:, 1:]

                temp_df.columns = temp_df.columns.astype(str).str.strip()

                temp_df = temp_df.reset_index(drop=True)

                revenue_tables[current_period] = temp_df

            # Extract period
            match = re.search(r'(FY \d{2}|Q[1-4] \d{2})', joined)

            if match:
                current_period = match.group(1)

            current_rows = []

        else:
            if not all(pd.isna(x) for x in row_values):
                current_rows.append(row_values)

    # Save final block
    if current_period and current_rows:
        temp_df = pd.DataFrame(current_rows)

        temp_df.columns = temp_df.iloc[0]
        temp_df = temp_df[1:]
        temp_df = temp_df.iloc[:, 1:]
        temp_df.columns = temp_df.columns.astype(str).str.strip()
        temp_df = temp_df.reset_index(drop=True)

        revenue_tables[current_period] = temp_df

    return revenue_tables
```

```python
def chunk_text(text, chunk_size=1000, overlap=200):
    """
    Split long text into overlapping chunks.
    """

    chunks = []

    start = 0

    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]

        if chunk.strip():
            chunks.append(chunk)

        start += chunk_size - overlap

    return chunks
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

# Embedding model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Chroma client
client = chromadb.Client()
```

```python
# STEP 2 FINAL — Chunk PDFs + Batch Embed + Store in Chroma

# Build chunked documents
chunked_documents = []

for doc in all_pdf_chunks:
    text_chunks = chunk_text(doc["content"])

    for i, chunk in enumerate(text_chunks):
        chunked_documents.append({
            "id": f"{doc['source']}-{doc['page']}-{i}",
            "content": chunk,
            "source": doc["source"],
            "page": doc["page"]
        })

print("Total chunks created:", len(chunked_documents))


# Prepare batch data
documents = [doc["content"] for doc in chunked_documents]

ids = [doc["id"] for doc in chunked_documents]

metadatas = [
    {
        "source": doc["source"],
        "page": doc["page"]
    }
    for doc in chunked_documents
]


# Generate embeddings in batch
embeddings = embedding_model.encode(
    documents,
    show_progress_bar=True
).tolist()


# Reset collection safely
try:
    client.delete_collection("infosys_financial_docs")
except:
    pass

collection = client.create_collection(
    name="infosys_financial_docs"
)


# Store everything
collection.add(
    ids=ids,
    embeddings=embeddings,
    documents=documents,
    metadatas=metadatas
)

print("Stored chunks in Chroma:", collection.count())
```

```python
def retrieve_documents(query, top_k=3):
    """
    Semantic document retrieval helper.
    """

    results = collection.query(
        query_texts=[query],
        n_results=top_k
    )

    return {
        "documents": results["documents"][0],
        "metadata": results["metadatas"][0]
    }
```

```python
def classify_query(query):
    """
    Intent classifier with correct priority ordering.
    """

    q = query.lower()

    # MOST COMPLEX FIRST
    if any(word in q for word in [
        "align",
        "relationship",
        "correlate",
        "impact",
        "versus earnings",
        "vs earnings"
    ]):
        return "MULTI_HOP"

    elif any(word in q for word in [
        "summary",
        "summarize",
        "overview",
        "report"
    ]):
        return "REPORT"

    elif any(word in q for word in [
        "compare",
        "difference",
        "growth",
        "margin",
        "change"
    ]):
        return "COMPARISON"

    elif any(word in q for word in [
        "risk",
        "strategy",
        "guidance",
        "outlook",
        "commentary",
        "cyber",
        "regulation"
    ]):
        return "QUALITATIVE"

    elif any(word in q for word in [
        "stock",
        "share price",
        "market",
        "returns",
        "volatility"
    ]):
        return "STOCK"

    else:
        return "LOOKUP"
```

```python
test_queries = [
    "What were the strategic risks?",
    "What was Q4 revenue?",
    "Compare Q2 and Q4 margin",
    "How did stock react after Q4 earnings?",
    "Summarize FY26 performance",
    "Did stock performance align with earnings?"
]

for q in test_queries:
    print(q, "->", classify_query(q))
```

```python
def route_query(query):
    """
    Route query to appropriate engine.
    """

    query_type = classify_query(query)

    routes = {
        "LOOKUP": "structured_financial_engine",
        "QUALITATIVE": "vector_retrieval_engine",
        "COMPARISON": "financial_analytics_engine",
        "STOCK": "stock_analytics_engine",
        "REPORT": "multi_source_report_engine",
        "MULTI_HOP": "hybrid_reasoning_engine"
    }

    return routes.get(query_type, "fallback_engine")
```

```python
for q in test_queries:
    print(q)
    print("Route:", route_query(q))
    print("------")
```

```python
def analyze_query(query):
    intent = classify_query(query)
    route = route_query(query)

    return {
        "query": query,
        "intent": intent,
        "route": route
    }
```

```python
def find_metric_row(df, metric_name):
    """
    Exact-first financial metric matching.
    Prevents wrong matches like Revenue vs Unearned Revenue.
    """

    first_col = df.columns[0]

    # Exact match first
    exact = df[
        df[first_col]
        .astype(str)
        .str.strip()
        .str.lower()
        == metric_name.lower()
    ]

    if not exact.empty:
        return exact.iloc[0]

    # Partial fallback
    partial = df[
        df[first_col]
        .astype(str)
        .str.strip()
        .str.lower()
        .str.contains(metric_name.lower(), na=False)
    ]

    if not partial.empty:
        return partial.iloc[0]

    return None
```

```python
def get_metric_value(df, metric_name, period):
    """
    Safe financial metric lookup with numeric conversion.
    """

    row = find_metric_row(df, metric_name)

    if row is None:
        return f"Metric '{metric_name}' not found"

    if period not in df.columns:
        return f"Period '{period}' not found"

    value = row[period]

    try:
        return float(value)
    except:
        return value
```

```python
def compare_metrics(df, metric_name, period1, period2):
    """
    Compare a financial metric across two periods safely.
    """

    val1 = get_metric_value(df, metric_name, period1)
    val2 = get_metric_value(df, metric_name, period2)

    if isinstance(val1, str) or isinstance(val2, str):
        return {
            "error": True,
            "message": f"{val1} | {val2}"
        }

    diff = val2 - val1

    pct_change = None

    if val1 != 0:
        pct_change = (diff / val1) * 100

    return {
        "metric": metric_name,
        "period1": period1,
        "value1": val1,
        "period2": period2,
        "value2": val2,
        "difference": diff,
        "pct_change": pct_change
    }
```

```python
def calculate_growth(df, metric_name, start_period, end_period):
    """
    Calculate growth percentage between two periods safely.
    """

    start_val = get_metric_value(df, metric_name, start_period)
    end_val = get_metric_value(df, metric_name, end_period)

    if isinstance(start_val, str) or isinstance(end_val, str):
        return {
            "error": True,
            "message": "Metric lookup failed"
        }

    if start_val == 0:
        return {
            "error": True,
            "message": "Division by zero"
        }

    growth = ((end_val - start_val) / start_val) * 100

    return {
        "metric": metric_name,
        "start_period": start_period,
        "end_period": end_period,
        "start_value": start_val,
        "end_value": end_val,
        "growth_pct": growth
    }
```

```python
def structured_financial_query(metric, period, sheet="profit_loss"):
    """
    Unified financial lookup.
    """

    sheet_map = {
        "profit_loss": profit_loss_df,
        "balance_sheet": balance_sheet_df,
        "cash_flow": cash_flow_df,
        "operating_metrics": operating_metrics_df
    }

    df = sheet_map[sheet]

    return get_metric_value(df, metric, period)
```
